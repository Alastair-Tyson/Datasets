```python
fake = Faker()
```

```python
def email_gen(fn, ln):
    roll = random.randint(1,6)
    if roll == 1:
        em = fn.lower()[0] + '.' + ln.lower() 
    elif roll == 2:
        em = fn.lower() + '.' + ln.lower()
    elif roll == 3:
        em = fn.lower() + '.' + ln.lower()[0:2]
    elif roll == 4:
        em = fn.lower() + ln.lower()
    elif roll == 5:
        em = random.choice(['darth', 'captain', 'super', 'admiral', 'general', 'awesome', 'tiny']) + ln.lower()
    else:
        em = fn.lower() + random.choice(['cheese', 'bear', 'cool', 'mountain', 'real'])
    return em + str(random.randint(1,999)) + '@' + random.choice(['aobell.com', 'coldmail.com', 'hmail.com', 'the' +ln.lower()+'s.com'])
```

```python
def get_state(date):
    if date < pd.to_datetime('14/10/2023', format = '%d/%m/%Y'):
        r = random.randint(1,10)
        if r < 2:
            output = 'New York'
        elif r < 5:
            output = random.choice(['New York', 'New Jersey'])
        elif r < 8:
            output = random.choice(['New York', 'Pennsylvania', 'New Jersey', 'Connecticut'])
        elif r < 10:
            output = random.choice(['New Hampshire', 'Massachusetts', 'Rhode Island', 'Ohio'])
        else:
            output = fake.state()
    else:
        r = random.randint(1,10)
        if r < 2:
            output = 'New York'
        elif r < 5:
            output = random.choice(['New York', 'New Jersey', 'Ohio'])
        elif r < 8:
            output = random.choice(['New York', 'Pennsylvania', 'New Jersey', 'Connecticut', 'Ohio'])
        elif r < 10:
            output = random.choice(['New Hampshire', 'Massachusetts', 'Rhode Island', 'Ohio'])
        else:
            output = fake.state()
    return output
```

```python
def gen_birth():
    month = random.randint(1,12)
    year_band = random.choices(population = [0, 1, 2, 3], weights = [0.5, 0.3, 0.15, 0.05], k = 1)[0]
    if year_band == 0:
        year = random.randint(1998, 2005)
    elif year_band == 1:
        year = random.randint(1988, 1997)
    elif year_band == 2:
        year = random.randint(1980, 1987)
    else:
        year = random.randint(1960, 1979)
    if month in [1, 3, 5, 7, 8, 10, 12]:
        day = random.randint(1, 31)
    elif month in [4, 6, 9, 11]:
        day = random.randint(1, 30)
    else:
        if year % 4 == 0:
            day = random.randint(1, 29)
        else: 
            day = random.randint(1, 28)
    return pd.to_datetime(str(day) +'/' + str(month) +'/' + str(year), format = '%d/%m/%Y')
def gen_start(year):
    month = random.randint(1,12)
    
    if month in [1, 3, 5, 7, 8, 10, 12]:
        day = random.randint(1, 31)
    elif month in [4, 6, 9, 11]:
        day = random.randint(1, 30)
    else:
        if year % 4 == 0:
            day = random.randint(1, 29)
        else: 
            day = random.randint(1, 28)
    return pd.to_datetime(str(day) +'/' + str(month) +'/' + str(year), format = '%d/%m/%Y')
```

```python
def customer_create(date):
    gen = random.randint(0,10)
    if gen > 3:
        fn = fake.first_name_female()
    else:
        fn = fake.first_name_male()
    ln = fake.last_name()
    email = email_gen(fn, ln)
    add = fake.address().strip().replace('\n',', ').split(', ')
    street = add[0]
    city = add[1]
    state = get_state(date)
    zipc = fake.zipcode()
    mail = random.choice([True, False])
    cid = fn.lower()[0] + ln.lower()[0:3] + str(random.randint(1000,9999))
    birthday = gen_birth()
    level = random.choices(population = ['Bronze', 'Silver', 'Gold'], weights = [0.5, 0.35, 0.15], k = 1)[0]
    member = gen_start(random.randint(2019, 2023))
    return cid, fn, ln, email, street, city, state, zipc, mail, birthday, level, member
```

```python
lines = ['Ocly', 'Verchecco', 'Piaris']
products = ['t-shirt', 'jeans','polo shirt', 'hoodie']
colors = ['grey', 'black', 'white']

price_l={'Ocly' : 0.95, 'Verchecco': 1.03, 'Piaris' : 1.01}
price_p = {'t-shirt': 12, 'jeans' : 35, 'polo shirt' : 15, 'hoodie' : 28}
```

```python
discounts = {'2f3' : 'buy two t-shirts, get third for half price', '10%pm' : '10% off jeans orders for members',
             '15%h' : '15% off hoodie if bought with a pair of jeans', 'odis' : '20% off Ocly for members'}
```

```python
product_dic = {}
prod_check = {}

for l in lines:
    for p in products:

        for c in colors:
                
                cost = round(price_p[p] * price_l[l], 2)
                if p != 'jeans' and p!= 'polo shirt':
                    pid = l.lower()[0] + p[0]  + '_' + c
                    product_dic[pid] = {'line' : l, 'category' : p, 'color' : c, 'rrp' : cost}
                    prod_check[pid] = {'line' : l, 'category' : p, 'color' : c}
                else:
                    pid = l.lower()[0] + p[0:4]
                    product_dic[pid] = {'line' : l, 'category' : p, 'color' : 'blue', 'rrp' : cost}
                    prod_check[pid] = {'line' : l, 'category' : p, 'color' : 'blue'}
```

```python
products_df = pd.DataFrame(product_dic).T
```

```python
stock_l = { 'Ocly' : 1.2, 'Verchecco': 0.65, 'Piaris' : 0.7}
stock_p = {'t-shirt': 400, 'jeans' : 300, 'polo shirt' : 300, 'hoodie' : 150}
stock_c = {'grey':1.05, 'black':1, 'white':0.8,'blue':0.9}
```

```python
pid_stock = {}
pid_target = {}

for k in product_dic.keys():
    info = product_dic[k]
    stock = int(round(stock_p[info['category']] * stock_l[info['line']] * stock_c[info['color']],2))
    pid_stock[k] = random.randint(round(stock * 0.85,0), round(stock * 1.05,0))
    pid_target[k] = round(product_dic[k]['rrp'] * stock * 0.95,2)
```

```python

cus_id = []
cus_fn = []
cus_ln = []
cus_email = []
cus_add1 = []
cus_add2 = []
cus_state = []
cus_level = []
cus_bir = []
cus_start = []
cus_zip = []
tran_ids = []
tid_count = 112923
tran_date = []
tran_cus = []
tran_stat = []
tran_price = []
tran_disc = []
start_date = pd.to_datetime('01/07/2023', format = '%d/%m/%Y')
tran_prod = []
tran_item =[]
del_id = []




while start_date <= pd.to_datetime('31/12/2023', format = '%d/%m/%Y'):
    base = 50
    if start_date.weekday() > 5:
        base += random.randint(15, 25)
    elif start_date == 5:
        base += random.randint(8, 12)
    else: 
        base -= random.randint(1, 10)
    if start_date.month > 10:
        base += random.randint(30, 45)
    elif start_date.month > 8:
        base += random.randint(22, 31)
    else:
        base += random.randint(-5, 6)
    no_purc = random.randint(base - 5, base + 5)
    for i in range(no_purc):
        guest = random.choices([True, False], weights = [0.6, 0.4], k = 1)[0]
        if guest == True:
            member_id = np.nan
            cus_info = customer_create(start_date)
            cus_id.append(member_id)
            cus_fn.append(cus_info[1])
            cus_ln.append(cus_info[2])
            cus_email.append(cus_info[3])
            cus_add1.append(cus_info[4])
            cus_add2.append(cus_info[5])
            cus_state.append(cus_info[6])
            cus_zip.append(cus_info[7])
            cus_level.append(cus_info[8])
            cus_bir.append(cus_info[9])
            cus_start.append(cus_info[10])
        else:
            cus_info = customer_create(start_date)
            member_id = cus_info[0]
            cus_id.append(cus_info[0])
            cus_fn.append(cus_info[1])
            cus_ln.append(cus_info[2])
            cus_email.append(cus_info[3])
            cus_add1.append(cus_info[4])
            cus_add2.append(cus_info[5])
            cus_state.append(cus_info[6])
            cus_zip.append(cus_info[7])
            cus_level.append(cus_info[8])
            cus_bir.append(cus_info[9])
            cus_start.append(cus_info[10])
        tran_id = 'AAA' + str(tid_count)
        del_id.append(tran_id)
        
        tid_count += 1
        
        no_purchases = random.choices(population = [1, 2, 3, 4, 5], weights = [0.6, 0.1, 0.25, 0.03, 0.02])[0]
        purchases = []
        jeans = False
        for p in range(no_purchases):
            if no_purchases >=3:
                prod_choice = random.choices(population = list(stock_l.keys()), weights = [0.35, 0.25, 0.30])[0]
                cat_choice = random.choices(population = list(stock_p.keys()), weights = [0.6, 0.1, 0.2, 0.15])[0]
            else:
                prod_choice = random.choices(population = list(stock_l.keys()), weights = [0.54, 0.3, 0.20])[0]
                cat_choice = random.choices(population = list(stock_p.keys()), weights = [0.45, 0.15, 0.2, 0.25])[0]
            if cat_choice != 'jeans' and col_choice != 'polo shirt':
                col_choice = random.choices(population = list(stock_c.keys()), weights = [0.3, 0.3, 0.3,0])[0]
            
            if cat_choice == 'jeans':
                jeans = True
                col_choice = 'blue'
            if cat_choice == 'polo shirt':
                col_choice = 'blue'
            purchases.append([prod_choice, col_choice, cat_choice])
        t_count = 0
        
        
        for s in range(len(purchases)):
            discounts_app = []
            disc = 1
            prod_pid = list(prod_check.keys())[list(prod_check.values()).index({'line': purchases[s][0], 'category': purchases[s][2], 'color': purchases[s][1]})]
            
            tran_ids.append(tran_id)
            tran_cus.append(member_id)
            tran_date.append(start_date)
            tran_prod.append(prod_pid)
            tran_item.append(s + 1)
            if jeans == True:
                if purchases[s][2] == 'hoodie' or purchases[s][2] == 'sweater':
                    disc = disc * 0.85
                    discounts_app.append('15%h')
            if guest == False:
                if purchases[s][2] == 'jeans':
                    disc = disc * 0.9
                    discounts_app.append('10%pm')
                
                if purchases[s][0] == 'Ocly':
                    disc = disc * 0.8
                    discounts_app.append('odis')
            if t_count == 2:
                disc = disc * 0.5
                discounts_app.append('2f3')
                t_count = 0
            else:
                if purchases[s][2] == 't-shirt':
                    t_count += 1
            
            tran_price.append(round(product_dic[prod_pid]['rrp'] * disc, 2))
            stri = ','
            for d in discounts_app:
                stri += ', '+ d
            if ',,' in stri:
                stri = stri.replace(',, ','')
            else:
                stri = np.nan
            tran_disc.append(stri)
            if start_date > pd.to_datetime('30/12/2023', format = '%d/%m/%Y'):
                tran_stat.append('Order Received')
                pid_stock[prod_pid] -= 1
            
                
                
            else:
                pid_stock[prod_pid] -= 1
                if pid_stock[prod_pid] > 0:
                    if start_date > pd.to_datetime('25/12/2023', format = '%d/%m/%Y'):
                        tran_stat.append('In Transit')
                    else:
                        tran_stat.append('Order Fulfilled')
                else:
                    tran_stat.append('Backlog')
    
                
                    
    start_date = start_date + datetime.timedelta(days=1)
```
