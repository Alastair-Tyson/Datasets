```python
def time_maker(date,time='All Day'):
    if time=='All Day':
        hour=random.randint(9,21)
    elif time=='Morning':
        hour=random.randint(8,13)
    elif time=='Afternoon':
        hour=random.randint(12,19)
    elif time=='Day':
        hour=random.randint(9,17)
    elif time=='Rest':
        r=random.randint(0,1)
        if r==0:
            hour=random.randint(7,8)
        else:
            hour=random.randint(17,21)
    else:
        hour=random.randint(20,23)
    year=date.year
    month=date.month
    day=date.day
    minute=random.randint(0,59)
    second=random.randint(0,59)
    return datetime.datetime(year,month,day,hour,minute,second)
```
