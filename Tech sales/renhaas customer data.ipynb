```python
import pandas as pd
import random
import numpy as np
```

```python
from faker import Faker
```

```python
fake = Faker('en_GB')
```

```python
fake.name()
```

```python
ids = []
titles = []
forename = []
surname = []
address_1 = []
address_2 = []
address_3 = []
county = []
postcode = []
telephone = []
age_band = []
package = []
monthly_sub = []
days_active = []
nps = []

start_id = random.randint(200, 500)

for i in range(random.randint(start_id, start_id + random.randint(500, 800))):
    name = fake.name()
    name_split = name.split(' ')
    if len(name_split) == 2:
        titles.append(np.nan)
        forename.append(name_split[0])
        surname.append(name_split[1])
    else:
        titles.append(name_split[0])
        forename.append(name_split[1])
        surname.append(name_split[2])
        
    add = fake.street_address().strip()
    add_split = add.split()
    if add_split[0] in ['Flat', 'Studio']:
        address_1.append(add_split[0] + ' ' + add_split[1])
        address_2.append(add_split[2] + ' ' + add_split[3][0].upper() + add_split[3][1:])
    else:
        address_1.append(add_split[0] + ' ' + add_split[1] + ' ' + add_split[2][0].upper() + add_split[2][1:])
        address_2.append(np.nan)

    
    address_3.append(fake.city())
    county.append(fake.county())
    postcode.append(fake.postcode())
    telephone.append(fake.phone_number())
    
    age = int(round(np.random.normal(loc = 45, scale = 10),0))
    age_band.append(age)
    pack = random.choices(population = ['basic', 'pro', 'enterprise'], weights = [0.5, 0.3, 0.2], k = 1)[0]
    package.append(pack)
    
    if pack == 'basic':
        sub = round(random.uniform(15, 20),2)
    elif pack == 'pro':
        sub = round(random.uniform(25, 35),2)
    else:
        sub = round(random.uniform(18, 40),2)
    monthly_sub.append(sub)
    days_act = np.random.normal(200, 40)
    days_active.append(int(round(days_act,0)))
    
    score = 0
    
    if pack == 'basic':
        score += random.randint(4,6)
    elif pack == 'pro':
        score += random.randint(8,12)
    else:
        score += random.randint(-2, 7)
        
    if days_act < 100:
        score += random.randint(-3,2)
    elif days_act < 150:
        score += random.randint(0,4)
    elif days_act < 200:
        score += random.randint(3,5)
    else:
        score += random.randint(8,12)
        
    if sub < 20:
        score += random.randint(8,12)
    elif sub < 30:
        random.randint(5,14)
    else:
        random.randint(2,20)
        
    if score < 10:
        nps.append(random.randint(1,6))
    elif score < 20:
        nps.append(random.randint(7,8))
    else:
        nps.append(random.randint(9,10))
    ids.append(start_id)
    start_id +=1
```

```python
df = pd.DataFrame({'customer_id' : ids, 'customer_title' : titles, 'forename' : forename, 'surname' : surname,
                   'address_1' : address_1, 'address_2' : address_2, 'address_3' : address_3, 'county' : county, 'postcode' : postcode,
                   'telephone' : telephone, 'age' : age_band, 'package_bought' : package, 'monthly_subscription' : monthly_sub,
                   'days_active' : days_active, 'recent_nps' : nps})
```
