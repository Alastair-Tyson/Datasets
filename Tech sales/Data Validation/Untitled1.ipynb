```python
import pandas as pd
from faker import Faker
import numpy as np
import random
```

```python
fake = Faker()
```

```python
error_types = {1:0.8, 2:0.15, 3:0.05}
```

```python
techs = []
for i in range(random.randint(5, 10)):
    techs.append(fake.first_name() + ' ' + fake.last_name())
    
teams = ['Dolphins', 'Cardinals', 'Bears']
tech_team = {}
for t in techs:
    team = random.choice(teams)
    tech_team[t] = team
```

```python
cus_names = []
accounts = []
start_dates = []
packages = []
emails = []
incs = []
resolve_date = []
stat = []
resolve = []
tech_ass = []
teamss = []

for i in range(random.randint(120, 220)):
    
    cus_fore = fake.first_name() 
    cus_las = fake.last_name()
 
    cus_account_number = cus_fore[0]+cus_las[0]+ str(random.randint(1000,9999))
    cus_start_date = random_date(pd.to_datetime('01/01/2018', format = '%d/%m/%Y'), pd.to_datetime('31/08/2023', format = '%d/%m/%Y'))
    package = random.choices(population = ['basic', 'pro', 'enterprise'], weights = [0.5, 0.2, 0.3], k =1)[0]
    acc_role = random.randint(1,10)
    if package == 'enterprise':
        cus_email = cus_fore.lower() + '.' + cus_las.lower() + '@' + fake.company().lower().replace('-','').replace(',','').replace(' ','') + random.choice(['.io','.org', 'com'])
    else:
        
        cus_email = cus_fore.lower() + '.' + cus_las.lower() + '@' + random.choice(['aobell.com', 'gsnail.com', 'coldmail.com']) 
    incident = random.choices(population = list(error_types.keys()), weights = list(error_types.values()), k=1)[0]
    inc_date = random_date_time(pd.to_datetime('01/09/2023', format = '%d/%m/%Y'), pd.to_datetime('30/09/2023', format = '%d/%m/%Y'))
    resolve_roll = random.randint(1,6)
    time_roll = random.randint(1, 100)
    if time_roll < 97:
        if resolve_roll > 4:
            res_date = np.nan
            status = 'Open'
        else:
            res_date = random_date_time(inc_date, pd.to_datetime('30/09/2023', format = '%d/%m/%Y'))
            status = 'Closed'
    else:
        res_date = random_date_time(pd.to_datetime('01/08/2023', format = '%d/%m/%Y'), pd.to_datetime('30/09/2023', format = '%d/%m/%Y'))
        status = 'Closed'
    null_roll = random.randint(1,100)
    if null_roll > 95:
        incident = np.nan
        res_date = np.nan
        status = 'Open'
    
    assigned = random.choice(techs)
    team = tech_team[assigned]
    
    
    
    cus_names.append(cus_fore + ' ' + cus_las)
    accounts.append(cus_account_number)
    start_dates.append(cus_start_date)
    packages.append(package)
    emails.append(cus_email)
    incs.append(incident)
    stat.append(str(inc_date) + '; ' + status)
    resolve.append(status)
    tech_ass.append(assigned)
    teamss.append(team)
    resolve_date.append(res_date)
    
    err_stat = False
    while err_stat == False:
        roll = random.randint(1, 50)
        if roll < 46:
            err_stat = True
        else:
            incident = random.choices(population = list(error_types.keys()), weights = list(error_types.values()), k=1)[0]
            inc_date = random_date_time(pd.to_datetime('01/09/2023', format = '%d/%m/%Y'), pd.to_datetime('30/09/2023', format = '%d/%m/%Y'))
            resolve_roll = random.randint(1,6)
            time_roll = random.randint(1, 100)
            if time_roll < 97:
                if resolve_roll > 4:
                    res_date = np.nan
                    status = 'Open'
                else:
                    res_date = random_date_time(inc_date, pd.to_datetime('30/09/2023', format = '%d/%m/%Y'))
                    status = 'Closed'
            cus_names.append(np.nan)#
            accounts.append(cus_account_number)#
            start_dates.append(np.nan)#
            packages.append(np.nan)#
            emails.append(np.nan)#
            incs.append(incident)#
            stat.append(str(inc_date) + '; ' + status)#
            resolve.append(status)
            tech_ass.append(assigned)
            teamss.append(team)
            resolve_date.append(res_date)
    
```

```python
df = pd.DataFrame({'date_of_incident;status' : stat, 'incident_type' : incs, 'resolved_date':resolve_date,
                   'technician_assigned':tech_ass, 'technician_team' : teamss,
                   'customer_number' : accounts, 'customer_name' : cus_names,
                  'customer_email':emails, 'account_created':start_dates, 'package':packages})
```

```python
def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```

```python
from random import randrange
from datetime import timedelta
def random_date_time(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```
