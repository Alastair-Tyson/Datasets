```python
for n in range(random.randint(3,6)):
    name = fake.first_name() + ' ' + fake.last_name()
    t = random.choice(list(teams.values()))
    teams[name] = t
```

```python
teams
```

```python
packs = []
purcs = []
vals = []
names = []
bizs = []
npss = []
con = []
nam = []
tea = []

for i in range(random.randint(100,120)):
    
    fn = fake.first_name()
    ln = fake.last_name()
    br = random.randint(1,6)
    
    if br > 4:
        biz = True
        pack = 'enterprise'
        comp = fake.company().replace(' ', '_').lower().replace('-','_').replace(',','').replace('and','').replace('__','_')
        em = fn.lower() + '.' + ln.lower() + '@' + comp + random.choice(['.com', '.io', '.org'])
    else:
        biz = False
        pack = random.choice(['basic', 'pro'])
        em = fn.lower() + '.' + ln.lower() + '@' +random.choice(['gsnail.com', 'coldmail.com', 'aobell.com'])
    year = 2023
    month = random.choices(population = ['06','07', '08', '09'], weights = [0.1, 0.3, 0.3, 0.3], k = 1)[0]
    day = random.randint(1, 30)
    purc = str(year) + '-' + month + '-' + str(day)
    sa = random.choice(list(teams.keys()))
    te = teams[sa]
    rt = random.randint(1, 100)
    if rt > 94:
        if te == 'MintCream':
            te = 'Mint Cream'
        elif te == 'LightGreen':
            te = 'Light Green'
        elif te == 'SaddleBrown':
            te = 'Saddle Brown'
        elif te == 'MediumBlue':
            te = 'Medium Blue'
        else:
            pass
    ph = fake.msisdn()
    no = ph[0:3] +'-'+ph[3:6]+'-'+ph[6:9]+'-'+ph[9:]
    
    if pack == 'basic':
        val = random.randint(130, 150)
        nps_cat = random.choices(population = list(nps_cats.keys()), weights = [0.2, 0.5, 0.3])[0]
    elif pack == 'pro':
        val = random.randint(220, 250)
        nps_cat = random.choices(population = list(nps_cats.keys()), weights = [0.4, 0.4, 0.2])[0]
    else:
        val = random.randint(300, 400)
        nps_cat = random.choices(population = list(nps_cats.keys()), weights = [0.1, 0.7, 0.2])[0]
    
    nps = random.randint(nps_cats[nps_cat][0], nps_cats[nps_cat][1])
    
    packs.append(pack)
    purcs.append(purc)
    vals.append(val)
    names.append(fn + ' ' + ln)
    bizs.append(biz)
    npss.append(nps)
    con.append(em +', ' + no)
    nam.append(sa)
    
    tea.append(te)
   
        
```

```python
q2 = pd.DataFrame({'package' : packs, 'purchase_date' : purcs, 'value' : vals, 'contact_name' : names,
                   'business_account' : bizs, 'NPS' : npss, 'contact' : con, 'name' : nam, 'team' : tea})
```
