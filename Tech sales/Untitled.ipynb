```python
import random
import numpy as np
import pandas as pd
from faker import Faker

fake = Faker()
```

```python
departments = ['People', 'Product', 'Marketing', 'GTM', 'Customer Support', 'Data and Insight']
```

```python

levels = ['Associate', 'Executive', 'Senior Executive', 'Principal']
roles = {}
for d in departments:
    print(d)
    number = int(input('how many roles? '))
    
    r = []
    for i in range(number):
        role = input('role? ')
        r.append(role)
    roles[d] = r
```

```python
roles['Data and Insight'] =['Data Scientist', 'Analyst']
```

```python
roles
```

```python
base_sal_dept = {}
for d in departments:
    print(d)
    base_sal_dept[d] = int(input('base? '))
```

```python
level_sal = {}
for l in levels:
    print(l)
    level_sal[l] = float(input('modifier? '))
```

```python
role_sal = {}
for r in roles.keys():
    for r2 in roles[r]:
        print(r2)
        role_sal[r2] = float(input('modifier?' ))
```

```python
reasons = ['Retired', 'Terminated','Resigned']
```

```python
people = {}
role_counter = {}
for r in roles.keys():
    for l in levels:
        role_counter[l + ' ' + r] = 0

for i in range(random.randint(120,180)):
    name = fake.first_name() + ' ' + fake.last_name()

    de = random.choice(departments)
    ro = random.choice(roles[de])
    status = False
    while status is False:
        lev = random.choices(population = levels, weights = [0.20, 0.55, 0.15, 0.1], k=1)[0]
        check = lev + ' ' + de
        current = role_counter[check]
        if current < level_limit[lev]:
            role_counter[check] += 1
            status = True
        else:
            pass
    people[name] = {'Role':ro, 'Department':de, 'Level':lev}
```

```python
level_limit = {'Associate' : 25, 'Executive' : 75, 'Senior Executive' : 10, 'Principal' : 5}
```

```python
from string import ascii_lowercase
```

```python
from random import randrange
from datetime import timedelta

def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```

```python
people = {}
role_counter = {}
for r in roles.keys():
    for l in levels:
        role_counter[l + ' ' + r] = 0

for i in range(random.randint(150,220)):
    name = fake.first_name() + ' ' + fake.last_name()

    de = random.choice(departments)
    ro = random.choice(roles[de])
    status = False
    while status is False:
        lev = random.choices(population = levels, weights = [0.20, 0.55, 0.15, 0.1], k=1)[0]
        check = lev + ' ' + de
        current = role_counter[check]
        if current < level_limit[lev]:
            role_counter[check] += 1
            status = True
        else:
            pass
    people[name] = {'Role':ro, 'Department':de, 'Level':lev}
emp_ids = []
name = []
emails = []
ro = []
dep = []
lev = []
country = []
start = []
end = []
tenure = []
reason = []
salary = []

for dude in people.keys():
    n = dude
    n_split = n.lower().split(' ')
    string = random.choices(population=ascii_lowercase, k = 3)
    emp_id = n_split[0][0]+n_split[1][0]+str(random.randint(100,999)) + string[0] + string[1] + string[2]
    email = n_split[0]+'.'+n_split[1]+'@'+'renhaas.org'
    count = random.choices(population = ['UK', 'US'], weights =[0.3, 0.7], k=1)[0]
    sal = random.randint(base_sal_dept[people[dude]['Department']],base_sal_dept[people[dude]['Department']]*1.02)
    sal2 = sal * random.uniform(role_sal[people[dude]['Role']],role_sal[people[dude]['Role']]*1.03)
    sal3 = round(sal2 * random.uniform(level_sal[people[dude]['Level']],level_sal[people[dude]['Level']]*1.03),0)
    roll_sal = random.randint(1, 100)
    if roll_sal > 96:
        sal3 += 100000
    if count == 'UK':
        sala = '£' + str(round(sal3 * 0.82,0))+'0'
    else:
        sala = '$' + str(sal3)+'0'
    start_date = random_date(pd.to_datetime('01/01/2010', format ='%d/%m/%Y'),pd.to_datetime('01/09/2023', format ='%d/%m/%Y'))
    start_string = str(start_date).split(' ')[0].split('-')
    st = start_string[2]+'/'+start_string[1]+'/'+start_string[0]
    roll = random.randint(0,100)
    if roll > 88:
        gone = 'Yes'
    else:
        gone = 'No'
    if gone == 'Yes':
        rea = random.choices(population = reasons, weights = [0.08, 0.05, 0.87], k=1)[0]
        end_date = random_date(start_date,pd.to_datetime('30/09/2023', format ='%d/%m/%Y'))
        end_string = str(end_date).split(' ')[0].split('-')
        en = end_string[2]+'/'+end_string[1]+'/'+end_string[0]
        delta = end_date - start_date
        ten = delta.days
        email =np.nan
    else:
        rea = np.nan
        en = np.nan
        delta = pd.to_datetime(date.today()) - start_date
        ten = delta.days
    if count == 'US':
        count = random.choice(['US', 'USA', 'United States'])
    if count == 'UK':
        count = random.choice(['UK', 'United Kingdom'])
        
    deps = people[dude]['Department']
    if deps == 'GTM':
        deps = random.choice(['GTM', 'Go To Market'])
    elif deps == 'Data and Insight':
        deps = random.choice(['Data and Insight' ,'D&I', 'Data'])
    rols = people[dude]['Role']
    if rols == 'Account Manager':
        rols = random.choice(['Account Manager', 'AM'])
    elif rols == 'Customer Success Manager':
        rols = random.choice(['CSM', 'Customer Success Manager'])
 
    emp_ids.append(emp_id)
    name.append(dude)
    emails.append(email)
    ro.append(rols)
    dep.append(deps)
    lev.append(people[dude]['Level'])
    country.append(count)
    start.append(st)
    end.append(en)
    tenure.append(ten)
    reason.append(rea)
    salary.append(sala)
    roll_dup = random.randint(1,100)
    if roll_dup > 95:
        emp_ids.append(emp_id)
        name.append(dude)
        emails.append(email)
        ro.append(rols)
        dep.append(deps)
        lev.append(people[dude]['Level'])
        country.append(count)
        start.append(st)
        end.append(en)
        tenure.append(ten)
        reason.append(rea)
        salary.append(sala)
        
```

```python
df=pd.DataFrame({'employee_id':emp_ids,'employee_name':name,'employee_email':emails,'role':ro,'level':lev,'Department':dep,
             'country':country, 'start_date':start,'end_date':end,' tenure':tenure,'reason_for_leaving':reason,'salary':salary})
```
