```python
import pandas as pd

loc = pd.read_csv('lu_daily_temps.csv')
```

```python
loc
```

```python
dates = []
lines = []
temp = []

for i in range(len(loc)):
    for j in ['Bakerloo', 'Central', 'Jubilee', 'Northern', 'Piccadilly',
       'Victoria', 'Waterloo_and_City', 'Sub-surface_lines']:
        if j == 'Sub-surface_lines':
            for k in ['District', 'Circle', 'Metropolitan', 'Hammersmith & City']:
                dates.append(loc.iloc[i,0])
                lines.append(k)
                temp.append(loc.loc[i,j])
            
        else:
            dates.append(loc.iloc[i,0])
            lines.append(j.replace('_and_', ' & '))
            temp.append(loc.loc[i,j])
```

```python
out = pd.DataFrame({'date' : dates, 'line': lines, 'temperature' : temp})
```

```python
out.to_csv('daily_temps.csv', index = False)
```

```python
df = pd.read_excel('lu_vis.xlsx',sheet_name = 'station_lines')
```

```python
stat = []
line = []

for i in range(len(df)):
    split = df.iloc[i,2].split(', ')
    for j in split:
        stat.append(df.iloc[i,0])
        line.append(j)
```

```python
lines = pd.DataFrame({'line_name':line, 'station_id' : stat})
```

```python
lines.to_csv('lines.csv', index = False)
```
