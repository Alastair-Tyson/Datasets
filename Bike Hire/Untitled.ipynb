```python
import random
```

```python
def birthday():
    age_band = random.choices([1960, 1970, 1980, 1990, 2000], [0.05, 0.1, 0.3, 0.4, 0.15])[0]
    if age_band == 1960:
        year = random.randint(1955, 1969)
    elif age_band == 2000:
        year = random.randint(2000, 2006)
    else:
        year = random.randint(age_band, age_band + 9)
    month = random.randint(1, 12)
    
    if month in [4, 6, 9, 11]:
        day = random.randint(1, 30)
    elif month in [1, 3, 5, 7, 8, 10, 12]:
        day = random.randint(1, 31)
    else:
        if year%4 == 0:
            day = random.randint(1, 29)
        else:
            day = random.randint(1, 28)
    return pd.to_datetime('{}/{}/{}'.format(year, month, day), format = '%Y/%m/%d')
```

```python
import numpy as np
```

```python
dob= []
for i in range(len(tt)):
    if tt.loc[i, 'member_casual']=='member':
        dob.append(birthday()) 
    else:
        dob.append(np.nan)
```

```python
tt['dob'] = dob
```

```python
import datetime
```

```python
s20 = []
e20 = []

for i in range(len(tt)):
    start = pd.to_datetime(tt.loc[i, 'started_at'])
    end = pd.to_datetime(tt.loc[i, 'ended_at'])
    
    s20.append(start + np.timedelta64(3,'Y'))
    e20.append(end + np.timedelta64(3,'Y'))
```

```python
tt['started_at'] = s20
tt['ended_at'] = e20
```
