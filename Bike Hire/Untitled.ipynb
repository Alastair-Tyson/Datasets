```python
def birthday():
    age_band = random.choices([1960, 1970, 1980, 1990, 2000], [0.05, 0.1, 0.3, 0.4, 0.15])[0]
    if age_band == 1960:
        year = random.randint(1955, 1969)
    elif age_band == 2000:
        year = random.randint(2000, 2006)
    else:
        year = random.randint(age_band, age_band + 9)
    month = random.randint(1, 12)
    
    if month in [4, 6, 9, 11]:
        day = random.randint(1, 30)
    elif month in [1, 3, 5, 7, 8, 10, 12]:
        day = random.randint(1, 31)
    else:
        if year%4 == 0:
            day = random.randint(1, 29)
        else:
            day = random.randint(1, 28)
    return pd.to_datetime('{}/{}/{}'.format(year, month, day), format = '%Y/%m/%d')
```

```python
def email(fn, sn):
    
    r = random.randint(1, 6)
    if r == 1:
        output = fn[0].lower() + sn.lower() 
    elif r == 2:
 
        output = fn.lower() + sn.lower() 
    elif r == 3:

        output = sn.lower() + fn.lower()
    elif r == 4:

        output = fn.lower() + '.' + sn.lower() 
    elif r == 5:

        output = fn.lower() + '_' + sn.lower() 
    else:
        output = ra.get_random_word()
        
    return output + '@' + fake.ascii_email().split('@')[1]
```

```python
fnames = []
snames = []
emails = []


for i in range(len(tn)):
    member = tn.loc[i, 'usertype']
    if member == 'Subscriber':
        r = np.random.normal(0, 1)
        if r < 1 or r > 1:
            n = random.choice(list(members.keys()))
            nsplit = n.split(' ')
            fnames.append(nsplit[0])
            snames.append(nsplit[1])
            emails.append(members[n])
        else:
            fn = fake.first_name()
            sn = fake.last_name()
            fnames.append(fn)
            snames.append(sn)
            e = email(fn, sn)
            emails.append(e)
            members[fn + ' ' + sn] = e
    else:
        fnames.append(np.nan)
        snames.append(np.nan)
        emails.append(np.nan)
```

```python
tn['first_name'] = fnames
tn['surname'] = snames
tn['email'] = emails
```
